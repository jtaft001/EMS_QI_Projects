```python
import json
import pandas as pd

# Use the exact path to your JSON file
json_path = "/mnt/nasdrive/shared/raw_data/home assistant/entities.json"
csv_path = "entities.csv"

# Load JSON data
with open(json_path, "r") as f:
    data = json.load(f)

# Extract fields
rows = []
for item in data:
    entity_id = item.get("entity_id", "")
    state = item.get("state", "")
    attributes = item.get("attributes", {})
    name = attributes.get("friendly_name", "")
    device_class = attributes.get("device_class", "")
    unit = attributes.get("unit_of_measurement", "")
    integration = attributes.get("integration", "")

    rows.append({
        "entity_id": entity_id,
        "name": name,
        "state": state,
        "device_class": device_class,
        "unit_of_measurement": unit,
        "integration": integration
    })

# Save to CSV
df = pd.DataFrame(rows)
df.to_csv(csv_path, index=False)

print(f"✅ Exported {len(df)} entities to:\n{csv_path}")
```
